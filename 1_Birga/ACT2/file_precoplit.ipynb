```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
IMOEX = pd.read_csv('IMOEX.csv')
BRENT = pd.read_csv('BRENT.csv')
SNP500 = pd.read_csv('SNP500.csv')
GOLD = pd.read_csv('GOLD.csv')
USDRUB = pd.read_csv('USDRUB.csv')
EURRUB = pd.read_csv('EUR_RUB.csv')
```

```python
IMOEX['year'] = IMOEX['<DATE>']//10000
IMOEX['month'] = (IMOEX['<DATE>'] - IMOEX['<DATE>']//10000 * 10000)//100
IMOEX['day'] = (IMOEX['<DATE>'] - IMOEX['<DATE>']//100 * 100)
IMOEX['date'] = pd.to_datetime(IMOEX['year'].astype(str) + ' ' + IMOEX['month'].astype(str)  + ' ' + IMOEX['day'].astype(str))
IMOEX.index = IMOEX['date']
IMOEX=IMOEX.drop(['date','<DATE>','<TIME>'],axis=1)
```

```python
IMOEX.to_csv('IMOEX')
IMOEX.head()
```
