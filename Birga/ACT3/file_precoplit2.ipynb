```python
BVSP['year'] = BVSP['<DATE>']//10000
BVSP['month'] = (BVSP['<DATE>'] - BVSP['<DATE>']//10000 * 10000)//100
BVSP['day'] = (BVSP['<DATE>'] - BVSP['<DATE>']//100 * 100)
BVSP['date'] = pd.to_datetime(BVSP['year'].astype(str) + ' ' + BVSP['month'].astype(str)  + ' ' + BVSP['day'].astype(str))
BVSP.index = BVSP['date']
BVSP=BVSP.drop(['date','<DATE>','<TIME>'],axis=1)
```

```python
BVSP.to_csv('BVSP')
BVSP.head()
```
