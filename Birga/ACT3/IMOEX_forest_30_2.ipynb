```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
data = pd.read_csv('data_imoex_30',index_col='date')
```

```python
ALUM = pd.read_csv('ALUM.csv', index_col='date')
BVSP = pd.read_csv('BVSP.csv', index_col='date')
FUTSEE100 = pd.read_csv('FUTSEE100.csv', index_col='date')
MEDIUM = pd.read_csv('MEDIUM.csv', index_col='date')
NATURALGAS = pd.read_csv('NATURALGAS.csv', index_col='date')
NICKEL = pd.read_csv('NICKEL.csv', index_col='date')
PALLADII = pd.read_csv('PALLADII.csv', index_col='date')
SEREBRO = pd.read_csv('SEREBRO.csv', index_col='date')
SHANGHAI = pd.read_csv('SHANGHAI.csv', index_col='date')
FIVETR = pd.read_csv('FIVETR.csv', index_col='date')
TWOTR= pd.read_csv('TWOTR.csv', index_col='date')
```

```python
ALUM.columns = [u'<OPEN>', u'<HIGH>', u'<LOW>', u'ALUM', u'<VOL>', u'year', u'month', u'day']
BVSP.columns = [u'<OPEN>', u'<HIGH>', u'<LOW>', u'BVSP', u'<VOL>', u'year', u'month', u'day']
FUTSEE100.columns = [u'<OPEN>', u'<HIGH>', u'<LOW>', u'FUTSEE100', u'<VOL>', u'year', u'month', u'day']
MEDIUM.columns = [u'<OPEN>', u'<HIGH>', u'<LOW>', u'MEDIUM', u'<VOL>', u'year', u'month', u'day']
NATURALGAS.columns = [u'<OPEN>', u'<HIGH>', u'<LOW>', u'NATURALGAS', u'<VOL>', u'year', u'month', u'day']
NICKEL.columns = [u'<OPEN>', u'<HIGH>', u'<LOW>', u'NICKEL', u'<VOL>', u'year', u'month', u'day']
PALLADII.columns = [u'<OPEN>', u'<HIGH>', u'<LOW>', u'PALLADII', u'<VOL>', u'year', u'month', u'day']
SEREBRO.columns = [u'<OPEN>', u'<HIGH>', u'<LOW>', u'SEREBRO', u'<VOL>', u'year', u'month', u'day']
SHANGHAI.columns = [u'<OPEN>', u'<HIGH>', u'<LOW>', u'SHANGHAI', u'<VOL>', u'year', u'month', u'day']
FIVETR.columns = [u'<OPEN>', u'<HIGH>', u'<LOW>', u'FIVETR', u'<VOL>', u'year', u'month', u'day']
TWOTR.columns = [u'<OPEN>', u'<HIGH>', u'<LOW>', u'TWOTR', u'<VOL>', u'year', u'month', u'day']
```

```python
data_2 = pd.merge(data, ALUM['ALUM'], left_index=True, right_index=True)
data_2 = pd.merge(data_2, BVSP['BVSP'], left_index=True, right_index=True)
data_2 = pd.merge(data_2, FUTSEE100['FUTSEE100'], left_index=True, right_index=True)
data_2 = pd.merge(data_2, NATURALGAS['NATURALGAS'], left_index=True, right_index=True)
data_2 = pd.merge(data_2, NICKEL['NICKEL'], left_index=True, right_index=True)
data_2 = pd.merge(data_2, PALLADII['PALLADII'], left_index=True, right_index=True)
data_2 = pd.merge(data_2, SEREBRO['SEREBRO'], left_index=True, right_index=True)
data_2 = pd.merge(data_2, SHANGHAI['SHANGHAI'], left_index=True, right_index=True)
data_2 = pd.merge(data_2, FIVETR['FIVETR'], left_index=True, right_index=True)
data_2 = pd.merge(data_2, TWOTR['TWOTR'], left_index=True, right_index=True)
data_2.head()
```

```python
data_2 = pd.read_csv('data_2_imoex_30',index_col='date')
```

```python
X = data_2.dropna().drop(['target_per','target_class'], axis=1)
y = data_2.dropna()['target_class']
```

```python
Xtrain = X[:1400]
ytrain = y[:1400]
Xtest = X[1400:]
ytest = y[1400:]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
#Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, shuffle = True, random_state=0)
model = RandomForestClassifier(n_estimators=1000, class_weight='balanced')
model.fit(Xtrain, ytrain)
ypred = model.predict(Xtest)
```

```python
from sklearn import metrics
print(metrics.classification_report(ypred, ytest))
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(ypred, ytest)
seaborn.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False)
plt.ylabel('true label')
plt.xlabel('predicted label')
```

```python
from sklearn.model_selection import learning_curve
from sklearn.utils import shuffle
model_2 = RandomForestClassifier(n_estimators=1000)
#X_sh, y_sh = shuffle(X, y, random_state=0)
train_sizes, train_scores, test_scores = learning_curve(model_2, X, y, train_sizes=np.arange(0.1,1,0.2),cv=3,scoring='accuracy')
```

```python
plt.plot(train_sizes, np.mean(train_scores, 1), color='blue', label='training score')
plt.plot(train_sizes, np.mean(test_scores, 1), color='red', label='validation score')
plt.ylim(0, 1.1)
plt.xlim(train_sizes[0], train_sizes[-1])
plt.xlabel('training size') # Размерность обучения
plt.ylabel('score')
plt.legend(loc='best')
```
